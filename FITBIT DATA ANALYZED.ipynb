```python
def remove_outliers_iqr(df, column, verbose=True):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    if verbose:
        removed = df.shape[0] - df_cleaned.shape[0]
        print(f"{column}: Removed {removed} outliers. Rows before: {df.shape[0]}, after: {df_cleaned.shape[0]}")
    
    return df_cleaned
```

```python
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'TotalSteps')
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'Calories')
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'VeryActiveMinutes')
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'FairlyActiveMinutes')
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'LightlyActiveMinutes')
df_dailyactivity = remove_outliers_iqr(df_dailyactivity, 'SedentaryMinutes')


df_weightlog = remove_outliers_iqr(df_weightlog, 'BMI')
df_weightlog = remove_outliers_iqr(df_weightlog, 'WeightKg')

df_hourlysteps = remove_outliers_iqr(df_hourlysteps, 'StepTotal')


df_hourlycalories = remove_outliers_iqr(df_hourlycalories, 'Calories')

if 'TotalIntensity' in df_hourlyintensities.columns:
    df_hourlyintensities = remove_outliers_iqr(df_hourlyintensities, 'TotalIntensity')
if 'AverageIntensity' in df_hourlyintensities.columns:
    df_hourlyintensities = remove_outliers_iqr(df_hourlyintensities, 'AverageIntensity')


if 'Value' in df_heartrate.columns:
    df_heartrate = remove_outliers_iqr(df_heartrate, 'Value')
```

```python
def segment_activity(steps):
    if steps < 3000:
        return 'Sedentary'
    elif steps <= 7000:
        return 'Moderately Active'
    else:
        return 'Highly Active'

avg_steps['ActivityLevel'] = avg_steps['TotalSteps'].apply(segment_activity)
```
